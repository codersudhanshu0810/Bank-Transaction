# Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# Loading Dataset

```python
df = pd.read_csv('bank_transactions.csv')
df
```

# Handling Null Values or isnull()

```python
df.isnull().sum()
```

```python
df.to_csv('bank_transactions.csv', index=False)
```

```python
df.info()
```

```python
df['CustomerDOB'] = df['CustomerDOB'].str.replace('/', '-')
df
```

```python
df['CustomerDOB'] = pd.to_datetime(df['CustomerDOB'], format='%d-%m-%Y')
df.info()
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], format='%d-%m-%Y')
df.info()
```

# Value Counts

```python
gdf = df['CustGender'].value_counts()
gdf = gdf.reset_index()
gdf
```

```python
plt.figure(figsize=(15,5))
plt.plot(gdf['CustGender'], gdf['count'])
plt.xlabel('Customer Gender')
plt.ylabel('Count')
plt.title('Count according to Gender')
plt.show()
```

```python
df
```

```python
ldf = df['CustLocation'].value_counts().head(15)
ldf = ldf.reset_index()
ldf
```

```python
plt.figure(figsize=(20,5))
plt.bar(ldf['CustLocation'], ldf['count'])
plt.xlabel('Customer Location')
plt.ylabel('Count')
plt.title('Count according to Location')
plt.savefig('Bank_Transaction.png', format='png')
```

```python
tdf = df.groupby('CustGender').agg(Transaction_Amount = ('TransactionAmount (INR)', 'sum'))
tdf = tdf.reset_index().sort_values(by='Transaction_Amount', ascending=False)
tdf
```

```python
plt.figure(figsize=(10,5))
plt.bar(tdf['CustGender'], tdf['Transaction_Amount'])
plt.xlabel('Customer Gender')
plt.ylabel('Transaction Amount')
plt.title('Transaction according to the Gender')
plt.savefig('Transaction according to Gender.png', format='png')
```

```python
ydf = df.groupby('CustLocation').agg(Transact_Amount = ('TransactionAmount (INR)', 'sum'))
ydf = ydf.reset_index().sort_values(by='Transact_Amount', ascending=False).head(10)
ydf
```

```python
plt.figure(figsize=(15,5))
plt.bar(ydf['CustLocation'], ydf['Transact_Amount'])
plt.xlabel('Customer Location')
plt.ylabel('Transaction Amount')
plt.title('Transaction according to the Location')
plt.savefig('Transaction according to Location.png', format='png')
```
